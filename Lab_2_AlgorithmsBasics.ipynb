```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
# It is use to give the no of times common element occurs in the given list.
import timeit
start_time=timeit.default_timer()
elapsed_time=(timeit.default_timer()-start_time)*1000
print("Elapsed time:",elapsed_time,"ms")
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
    even_cnt = 0
    odd_cnt = 0
    for num in arr:
        if num % 2 == 0:
            even_cnt += 1
        else:
            odd_cnt += 1

    return even_cnt, odd_cnt
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr, target):
    if len(arr) == 0:
        return -1
    for index, value in enumerate(arr):
        if value == target:
            return index
    return -1
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
print(linear_Search([1,3,5,7,9],5))
import timeit
start_time=timeit.default_timer()
elapsed_time=(timeit.default_timer()-start_time)*1000
print("Elapsed time:",elapsed_time,"ms")
```

```python
# write a function in python named as Binary_Search to perform binary search in 1D-array in unsorted array
def Binary_Search(arr, target):
  arr.sort()
  left, right = 0, len(arr) - 1
  while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

  return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
print(Binary_Search([1, 3, 5, 7, 9,15,16,14,45],14))
import timeit
start_time=timeit.default_timer()
elapsed_time=(timeit.default_timer()-start_time)*1000
print("Elapsed time:",elapsed_time,"ms")
```
